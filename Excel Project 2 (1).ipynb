# Creating The Master Dataset
--> after filtering all dataset is Master dataset
--> we use the Concate function in Pandas to Two different dataset

```python
master_data = pd.concat([merged_data, filtered_3], ignore_index=True)
master_data
```

```python
# Find the Skewness, kutosis and the Outliers of the data
--> use the Numpy and Scipy stats the different calculation, Matplotlib and Seaborn making the plot
--> kewness, Kurtosis and Outliers calculated for all columns
--> For loop is used for iterating all data
--> Box plot is used for outliers
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
 
# Numerical columns to analyze
num_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
 
# Loop through each numerical column
for col in num_cols:
    data = master_data[col].dropna()
    # Skewness and Kurtosis
    sk = skew(data)
    kt = kurtosis(data)
    print(f"\nColumn: {col}")
    print(f"  Skewness: {sk:.4f}")
    print(f"  Kurtosis: {kt:.4f}")
    # IQR and Outlier Detection
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    print(f"  IQR: {IQR:.4f}")
    print(f"  Number of outliers: {outliers.count()}")
    # Boxplot
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

# Heatmap is the best for find corelation and corfficient between two set
--> use Average temparature and weekday during whole season

```python
pivot_temp = master_data.pivot_table(index='weekday', columns='hr', values='temp', aggfunc='mean')
 
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_temp, cmap="coolwarm", annot=False)

plt.title("Average Temperature by Hour and Weekday")

plt.xlabel("Hour")

plt.ylabel("Weekday (Starting from Sunday as 0)")

plt.show()
```
